```python
import pandas as pd
import numpy as np
import numpy as np
import os
import glob
```

```python
path = '/work/users/s/m/smerrill/SymMV'
```

### resnet

```python
# Get list of all .npy files in the current directory
resnet_path = os.path.join(path, 'resnet/resnet101')
resnet_files = os.listdir(resnet_path)

for file in resnet_files:
    if file.endswith('_fps.npy') or file.endswith('_timestamps_ms.npy'):
        print(f"Deleting: {file}")
        os.remove(os.path.join(resnet_path, file))
    elif file.endswith('_resnet.npy'):
        new_name = file.replace('_resnet.npy', '.npy')
        print(f"Renaming: {file} -> {new_name}")
        os.rename(os.path.join(resnet_path, file), os.path.join(resnet_path, new_name))
```

### clip

```python
# Get list of all .npy files in the current directory
clip_path = os.path.join(path, 'clip/ViT-B_32')
clip_files = os.listdir(clip_path)

for file in clip_files:
    if file.endswith('_fps.npy') or file.endswith('_timestamps_ms.npy'):
        print(f"Deleting: {file}")
        os.remove(os.path.join(clip_path, file))
    elif file.endswith('_clip.npy'):
        new_name = file.replace('_clip.npy', '.npy')
        print(f"Renaming: {file} -> {new_name}")
        os.rename(os.path.join(clip_path, file), os.path.join(clip_path, new_name))
```

### i3d

```python
i3d_path = os.path.join(path, 'i3d')
i3d_files = os.listdir(i3d_path)
vids = [file.split('_flow')[0] for file in i3d_files if 'flow' in file]
```

```python
for vid in vids:
    try:
        flow_feats = np.load(os.path.join(path, 'i3d', f'{vid}_flow.npy'))
        rgb_feats = np.load(os.path.join(path, 'i3d', f'{vid}_rgb.npy'))

        combined_feats = np.hstack([flow_feats, rgb_feats])
        np.save(os.path.join(path, 'i3d', f'{vid}.npy'), combined_feats)
        print(f"VID {vid} Processed Successfully")
    except Exception as e:
        print(e)
```
